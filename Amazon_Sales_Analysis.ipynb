```python
import pandas as pd
from matplotlib import pyplot as plt 
import datetime as dt
import seaborn as sns
```

```python
# Read the csv file
df = pd.read_csv('C:/Users/ejjid/Downloads/Amazon Sales data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x = df['Order Year'], y = df['Total Revenue'], color = 'green')
plt.title("Sales Performance Analysis: Trends in Units Sold (2010-2017)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x = df['Order Year'], y = df['Total Profit'], color='green')
plt.title('Profit Trends Over 2010-2017: An Analysis')
plt.show()
```

```python
data_preproc = pd.DataFrame({
    'Year': df['Order Year'], 
    'Total Revenue': df['Total Revenue'],
    'Total Cost' : df['Total Cost']
    })
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='value' , hue='variable', data=pd.melt(data_preproc, ['Year']), errorbar=None)
plt.legend()
plt.grid()
plt.title("Total Revenue vs Total Cost 2010-2017")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df, x = df['Order Month'], y = df['Total Profit'])
plt.title("Monthly Total Profit: A Box Plot Analysis (2010-2017)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x = df['Order Month'], y = df['Total Revenue'])
plt.title('Sales By Month 2010-2017')
plt.show()
```

```python
df_2010 = df[df['Order Year']==2010]
```

```python
df_2010.head()
```

```python
df_2010 = df_2010.sort_values(by='Total Profit', ascending = False)
```

```python
df_2010.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Profit'])
plt.grid()
plt.title("Monthly Profit Trends in 2010: A Bar Plot Analysis")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Revenue'], color = 'green')
plt.grid()
plt.title("Monthly Revenue Trends in 2010: A Line Plot Analysis")
plt.show()
```

```python
df.sort_values(by = 'Total Profit', ascending=False)
df[df['Country']=='Turkmenistan']
```

```python
CountryProfit = df.groupby('Country')['Total Profit'].sum().reset_index()
CountryProfit = CountryProfit.sort_values(by = 'Total Profit', ascending=False).reset_index()
CountryProfit.head()
```

```python
plt.figure(figsize=(14, 14))
sns.set_palette("pastel")
sns.set(style="whitegrid")
plt.title('Total Profit by Country')
sns.barplot(x='Total Profit', y='Country', data=CountryProfit, errorbar=None, hue = 'Total Profit')
plt.show()
```

```python
ProductProfit = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False).reset_index()
ProductProfit.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = ProductProfit, x = 'Total Profit', y = 'Item Type', hue = 'Total Profit')
plt.title("Profit by Item Type")
plt.show()
```
